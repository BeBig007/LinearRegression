```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "data/data.csv"
data= pd.read_csv(path)

data.head()

x = data['km']
y = data['price']
```

```python
plt.figure()
plt.scatter(x, y)
plt.xlabel('km')
plt.ylabel('price')
plt.show()
```

```python
def estimate_coef(x, y):
    # Calculating Number of Observations
    n = np.size(x)
    
    # Calculating Means
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    # Calculating Cross-Deviation and Deviation about x
    ss_xy = np.sum(y * x) - n * mean_x * mean_y
    ss_xx = np.sum(x * x) - n * mean_x * mean_x
    
    # Calculating Regression Coefficients
    b_1 = ss_xy / ss_xx
    b_0 = mean_y - b_1 * mean_x
    
    return b_0, b_1

b = estimate_coef(data['km'], data['price'])
```

```python
def plot_regression_line(x, y, b):
    plt.figure()
    plt.scatter(x, y)
    
    y_pred = b[0] + b[1] * x
    plt.plot(x, y_pred, color="red")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

plot_regression_line(x, y, b)
```
